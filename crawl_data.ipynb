### 1.2. Tiến hành lấy dữ liệu

```python
stocks_list = ['AGR', 'BID', 'VCB', 'CTG', 'TPB', 'HDB', 'VPB']  

# Tạo thư mục lưu dữ liệu nếu chưa có
output_folder = 'stock_data_output'
os.makedirs(output_folder, exist_ok=True)

# Dictionary để lưu dữ liệu cổ phiếu
stock_data = {}

# Lặp qua từng mã cổ phiếu và tải dữ liệu
for stock_symbol in stocks_list:
    try:
        # Tạo đối tượng Quote và lấy dữ liệu lịch sử
        quote = Quote(symbol=stock_symbol, source='VCI')
        data = quote.history(start='2017-01-01', end='2025-03-23', interval='1D')

        if data is not None and not data.empty:  # Kiểm tra dữ liệu hợp lệ
            stock_data[stock_symbol] = data
            data.to_csv(f'{output_folder}/{stock_symbol}.csv', index=True)
            print(f"Đã lưu dữ liệu: {stock_symbol}")
        else:
            print(f"Không có dữ liệu cho: {stock_symbol}")

    except Exception as e:
        print(f"Lỗi khi lấy dữ liệu {stock_symbol}: {e}")
```
