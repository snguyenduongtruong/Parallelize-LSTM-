### 1.2. Ti·∫øn h√†nh l·∫•y d·ªØ li·ªáu

```python
stocks_list = ['AGR', 'BID', 'VCB', 'CTG', 'TPB', 'HDB', 'VPB']  

# T·∫°o th∆∞ m·ª•c l∆∞u d·ªØ li·ªáu n·∫øu ch∆∞a c√≥
output_folder = 'stock_data_output'
os.makedirs(output_folder, exist_ok=True)

# Dictionary ƒë·ªÉ l∆∞u d·ªØ li·ªáu c·ªï phi·∫øu
stock_data = {}

# L·∫∑p qua t·ª´ng m√£ c·ªï phi·∫øu v√† t·∫£i d·ªØ li·ªáu
for stock_symbol in stocks_list:
    try:
        # T·∫°o ƒë·ªëi t∆∞·ª£ng Quote v√† l·∫•y d·ªØ li·ªáu l·ªãch s·ª≠
        quote = Quote(symbol=stock_symbol, source='VCI')
        data = quote.history(start='2017-01-01', end='2025-03-23', interval='1D')

        if data is not None and not data.empty:  # Ki·ªÉm tra d·ªØ li·ªáu h·ª£p l·ªá
            stock_data[stock_symbol] = data
            data.to_csv(f'{output_folder}/{stock_symbol}.csv', index=True)
            print(f"ƒê√£ l∆∞u d·ªØ li·ªáu: {stock_symbol}")
        else:
            print(f"Kh√¥ng c√≥ d·ªØ li·ªáu cho: {stock_symbol}")

    except Exception as e:
        print(f"L·ªói khi l·∫•y d·ªØ li·ªáu {stock_symbol}: {e}")
```
